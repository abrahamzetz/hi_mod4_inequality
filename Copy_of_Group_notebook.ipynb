Abraham & Joyce - Percentage of job distribution per city per role compared to the total average per role

````python
#Abraham & Joyce - Last one: Percentage of job distribution per city per role compared to national percentage

pd.read_sql(
  """
  with jobs_per_role as (
       select 
          role_id,
          count(created_at) as jobs_role
       FROM jobs
       group by role_id
       having role_id is not null
       ),

        totaljobs as (
          select
              count(created_at) as total_jobs
              from jobs
        ),


        national_job_percentage as (
          select
          role_id,
          roles.title
          jobs_role,
          (100.0*jobs_role/total_jobs) as percentage
          from jobs_per_role, totaljobs
          join roles
            on roles.id = jobs_per_role.role_id
          group by role_id
          ),


        jobs_per_city as (
            select
                city_id,
                count(j.created_at) as total_jobs
            from jobs j
            join employers e
              on j.employer_id = e.id
            group by city_id
          )



    select 
      roles.title,
      cities.name,
      --jobs_per_city.total_jobs,
      --count(jobs.created_at) as total_jobs_role,
      round((100.0*count(jobs.created_at)/jobs_per_city.total_jobs),2) as percentage_city,
      round(national_job_percentage.percentage, 2) as percentage_national,
      round(((100.0*count(jobs.created_at)/jobs_per_city.total_jobs)-national_job_percentage.percentage), 2) as difference
    from jobs
    join employers
      on jobs.employer_id = employers.id 
    join jobs_per_city
      on jobs_per_city.city_id = employers.city_id
    join national_job_percentage
      on national_job_percentage.role_id = jobs.role_id
    join roles
      on roles.id = jobs.role_id
    join cities
      on cities.id = employers.city_id
    group by employers.city_id,jobs.role_id






       
  """, db_conn)
````

Jonatan - **Average payout per role nationally** and average payout per role per city

````python
pd.read_sql(
  """
     -- Average hourly payrate
      WITH nat_avg_payrate AS(
        SELECT
          r.title,
          ROUND(avg(j.pay_rate), 2) AS avg_hourly_payrate,
          ROUND(avg(j.duration_hours * j.pay_rate), 2) AS avg_pay_rate
       FROM jobs j
       JOIN roles r ON j.role_id = r.id
       GROUP BY j.role_id
       ),
      city_avg_pay AS(
        SELECT 
            cities.name as City, 
            roles.title as JobTitle, 
            count(jobs.created_at) AS JobCount,
            ROUND(AVG(jobs.pay_rate),3) as AveragePayRate, 
            ROUND(AVG(jobs.pay_rate*jobs.duration_hours)) as AverageSumPaid
       FROM jobs
       JOIN cities ON employers.city_id = cities.id
       JOIN employers ON jobs.employer_id = employers.id
       JOIN roles ON jobs.role_id = roles.id
       GROUP BY employers.city_id, jobs.role_id
       )

      SELECT 
          c.City as City,
          c.JobTitle AS Jobtitle,
          c.AveragePayRate AS Regionalaveragehourlypayrate,
          n.avg_hourly_payrate AS Nationalaveragehourlypayrate,
          c.AverageSumPaid AS RegionalSumPaid,
          n.avg_pay_rate AS NationalSumPaid,
          ROUND((c.AveragePayRate - n.avg_hourly_payrate), 2) AS Difference,
          ROUND((c.AverageSumPaid - n.avg_pay_rate), 2) AS SumDifference,
          c.JobCount
      FROM city_avg_pay c
      JOIN nat_avg_payrate n ON c.JobTitle = n.title
      GROUP BY City, JobTitle
      ORDER BY SumDifference DESC

  """, db_conn)
````

Axel - Average payout per role nationally and average payout per role per city

````python
pd.read_sql(
  """
     -- Average hourly payrate
      WITH nat_avg_payrate AS(
        SELECT
          r.title,
          ROUND(avg(j.pay_rate), 2) AS avg_hourly_payrate,
          ROUND(avg(j.duration_hours * j.pay_rate), 2) AS avg_pay_rate
       FROM jobs j
       JOIN roles r ON j.role_id = r.id
       GROUP BY j.role_id
       ),
      city_avg_pay AS(
        SELECT 
            cities.name as City, 
            roles.title as JobTitle, 
            count(jobs.created_at) AS JobCount,
            ROUND(AVG(jobs.pay_rate),3) as AveragePayRate, 
            ROUND(AVG(jobs.pay_rate*jobs.duration_hours)) as AverageSumPaid
       FROM jobs
       JOIN cities ON employers.city_id = cities.id
       JOIN employers ON jobs.employer_id = employers.id
       JOIN roles ON jobs.role_id = roles.id
       GROUP BY employers.city_id, jobs.role_id
       ),
      Difference AS( 
      SELECT 
          c.City as City,
          c.JobTitle AS Jobtitle,
          c.AveragePayRate AS Regionalaveragehourlypayrate,
          n.avg_hourly_payrate AS Nationalaveragehourlypayrate,
          c.AverageSumPaid AS RegionalSumPaid,
          n.avg_pay_rate AS NationalSumPaid,
          ROUND((c.AveragePayRate - n.avg_hourly_payrate), 2) AS Difference,
          ROUND((c.AverageSumPaid - n.avg_pay_rate), 2) AS SumDifference,
          c.JobCount
      FROM city_avg_pay c
      JOIN nat_avg_payrate n ON c.JobTitle = n.title
      GROUP BY City, JobTitle
      ORDER BY JobCount DESC
      ),
      temp AS(SELECT 
          City,
          JobTitle,
          Regionalaveragehourlypayrate as RegHourPay,
          Nationalaveragehourlypayrate as NatHourPay,
          Difference,
          RegionalSumPaid,
          NationalSumPaid,
          SumDifference,
          JobCount
      FROM Difference
      --WHERE ABS(Difference) >= 0.10
      WHERE JobCount >= 50
      GROUP BY City, JobTitle
      ORDER BY SumDifference DESC)

      SELECT 
          City, 
          JobTitle, 
          RegHourPay, 
          NatHourPay, 
          MAX(Difference) as 'Hourly Difference', 
          RegionalSumPaid, 
          NationalSumPaid, 
          JobCount, 
          MAX(SumDifference) as 'Total Difference'
      from temp
      GROUP BY City
  """, db_conn)
````

Iraci - Completion rate of jobs by city and role in %

````python
pd.read_sql(
    """
    With table_totals as (
      select
        c.name as city,
        count(j.booked_worker) as booked_jobs,
        count(j.id) as total_offers
    from jobs as j
    join employers as e on j.employer_id = e.id
    join cities as c on e.city_id = c.id
    group by city
   )
    select
      c.name ,
      r.title,
      count(j.id) as job_by_role,
      table_totals.booked_jobs,
     (100.0*count(j.id)/booked_jobs) as percentage
    from jobs as j
    join employers as e on j.employer_id = e.id
    join cities as c on e.city_id = c.id
    join roles as r on j.role_id = r.id
    join table_totals on c.name = table_totals.city
    group by c.name, r.title
    """, db_conn)

````
